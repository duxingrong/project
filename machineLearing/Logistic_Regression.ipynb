Sigmoid Function = 1/(1+e**(-z)) 这就是逻辑回归的模型

```python
#导入数据
import pandas as pd
url = "Diabetes_Data.csv"
data = pd.read_csv(url)
data 
```

```python
#资料处理
data['Gender'] =  data["Gender"].map({"男生":1 , "女生": 0 })
data
```

```python
#训练集和测试集
#资料预处理 ：训练集，测试集
y = data["Diabetes"]
x = data[["Age" ,"Weight","BloodSugar","Gender"]]
#分成测试和训练
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=0.2,random_state=87)

#转换成numpy,方便计算
x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
x_test = x_test.to_numpy()
y_test = y_test.to_numpy()
```

```python
#特征缩放 Feature Scaling  可以加速gradient_descent 
# 标准化 standardization
# x-mean(x)
# ---------
# x标准差

#用库函数来做标准化
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
#直接用训练集的标准差和平均值来标准化测试集合
x_test = scaler.transform(x_test)
```

```python
import numpy as np
def sigmoid(z):
    return  1/(1+np.exp(-z))
    
```

```python

w  = np.array([1,2,3,4])
b = 1 
z  = (w*x_train).sum(axis=1)+b
sigmoid(z)
```
